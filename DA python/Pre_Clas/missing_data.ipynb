## Filling Null Values

```python
df3 = df.fillna(value=5)
df3.head()
```

```python
df4 = df.fillna(method = "pad")
df4.head()
```

```python
df4.isnull().sum()
```

```python
df5 = df.fillna(method="bfill")
df5.head()
```

```python
df5.isnull().sum()
```

```python
df6 = df.fillna(method="pad",axis=1)
df6.head()
```

```python
df6.isnull().sum()
```

```python
df7 = df.fillna({"Notes": "abcd", "Status": 1234})
df7.head()
```

```python
df9 = df.fillna(value=df["TotalPay"].mean())
df9.head()
```

```python
df10 = df.fillna(value=df["TotalPay"].min())
df10.head()
```

## dropna 

```python
df11 = df.dropna(how="all")
df11.head()
```

## replace

```python
df12 = df.replace(to_replace=np.nan, value=875465)
df12.head()
```

```python
df13 = df.replace(to_replace=0.00, value=11.11)
df13.head()
```

## interpolate()

```python
df["BasePay"] = df["BasePay"].interpolate(method="linear")
df
```

```python
df["BasePay"] = df["BasePay"].interpolate(method="linear",limit_direction="forward")
df
```
