```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "name": ["Sami","Kezban","Elvan","Vildan"],
    "yaş":[41,39,8,12]
})
df
```

```python
pd.get_dummies(df, columns=["name"])
```

```python
pd.get_dummies(df, columns=["name"], prefix="Dmy")
```

```python
pd.get_dummies(df, columns=["name"], prefix="Dmy", prefix_sep="*")
```

```python
pd.get_dummies(df, columns=["name"], drop_first=True)
```
