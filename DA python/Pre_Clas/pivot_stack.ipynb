```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
weather = {
    "date": ["5/1/2017","5/2/2017","5/3/2017","5/1/2017","5/2/2017","5/3/2017","5/1/2017","5/2/2017","5/3/2017"],
    "city": ["new york","new york","new york","mumbai","mumbai","mumbai","beijing","beijing","beijing",],
    "tempereture": [65,66,68,75,78,82,80,77,79],
    "humudity": [56,58,60,80,83,85,26,30,35]
}
df = pd.DataFrame(weather)
df
```

```python
df.pivot(index= "date", columns="city",values="humudity")
```

```python
df.pivot(index= "date", columns="city",values=["humudity","tempereture"])
```

```python
df.pivot(index="humudity",columns="city")
```

```python
df2 = pd.DataFrame({
    "date": ["5/1/2017","5/1/2017","5/2/2017","5/2/2017","5/1/2017","5/1/2017","5/2/2017","5/2/2017"],
    "city": ["new york","new york","new york","new york","mumbai","mumbai","mumbai","mumbai"],
    "temperature": [65,61,70,72,75,78,82,80],
    "humudity": [56,54,60,62,80,83,85,26]
})
df2
```

```python
df_single_level_cols = pd.DataFrame([[0, 1], [2, 3]],
                                    index=['cat', 'dog'],
                                    columns=['weight', 'height'])
df_single_level_cols
```

```python
df_single_level_cols.stack()
```

```python
multicol1 = pd.MultiIndex.from_tuples([('weight', 'kg'),
                                       ('weight', 'pounds')])
df_multi_level_cols1 = pd.DataFrame([[1, 2], [2, 4]],
                                    index=['cat', 'dog'],
                                    columns=multicol1)
df_multi_level_cols1
```

```python
df_multi_level_cols1.stack()
```

```python
df_multi_level_cols1.stack(level=0)
```

```python
df_multi_level_cols1.stack(level=1)
```

```python
multicol2 = pd.MultiIndex.from_tuples([('weight', 'kg'),
                                       ('height', 'm')])
df_multi_level_cols2 = pd.DataFrame([[1.0, 2.0], [3.0, 4.0]],
                                    index=['cat', 'dog'],
                                    columns=multicol2)
df_multi_level_cols2
```

```python
df_multi_level_cols2.stack()
```

```python
df_multi_level_cols2.stack(0)
```
